```python
# Later, load in the results for analysis
with open('d_baseline.p', 'rb') as f:
    base = pickle.load(f)
print(len(base['S6']['d_baseline'].keys()))

del base['S6']['d_baseline'][None]
```

Need to get a table, such that each column is length of mutation (3/-6/-9) and each row is position: how many are present.

```python
def counts_per_lib(base):
    dna_len_count = {3: 0, 6:0, 9:0}
    for prot_mutation, entry in base['S6']['d_baseline'].items():
        for dna_mutation, count in entry['dna'].items():
            try:
                if dna_mutation is None:
                    print(prot_mutation, dna_mutation, count)
                    continue
                elif dna_mutation == ():
                    continue
                
                else:       
                    dna_len = int(dna_mutation[0][-1])
                    dna_len_count[dna_len] += 1
            except IndexError:
                print(prot_mutation, dna_mutation, count)
                raise
    print(dna_len_count)

counts_per_lib(base)
```
